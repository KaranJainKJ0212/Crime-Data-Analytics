```python
# ðŸ“Š District-Wise Crime Analysis Project (2001â€“2014)
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Display more rows/cols in Kaggle
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# Step 1: Load and Preprocess the Data
df = pd.read_excel('/kaggle/input/crime-data/Crime_Data_District_Wise.xlsx', sheet_name='Sheet1')

# Clean and convert
df = df[pd.to_numeric(df['YEAR'], errors='coerce').notnull()]
df['YEAR'] = df['YEAR'].astype(int)
df.drop(columns=['Unnamed: 0'], errors='ignore', inplace=True)

# Step 1.5: Standardize DISTRICT names to uppercase
df['DISTRICT'] = df['DISTRICT'].str.upper()

# Convert crime columns to numeric
crime_columns = df.columns.difference(['STATE/UT', 'DISTRICT', 'YEAR'])
df[crime_columns] = df[crime_columns].apply(pd.to_numeric, errors='coerce')

# Step 2: Descriptive Statistics by District
stats = df.groupby('DISTRICT')[crime_columns].agg(['mean', 'std'])
stats.columns = ['_'.join(col) for col in stats.columns]
stats.reset_index(inplace=True)
display(stats.head())

# Step 3: PCA on Mean Crime Types
mean_vals = stats.filter(regex='_mean$')
scaled = StandardScaler().fit_transform(mean_vals)
pca = PCA(n_components=3)
pca_features = pca.fit_transform(scaled)

# Print explained variance
print("Explained Variance Ratios:", pca.explained_variance_ratio_)

# Show PCA loadings
loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(3)], index=mean_vals.columns)
display(loadings)

# Step 4: Trend Analysis (Regression)
trends = []
for district, group in df.groupby('DISTRICT'):
    X = group[['YEAR']]
    y = group['TOTAL IPC CRIMES']
    if y.notnull().all() and len(y) > 1:
        model = LinearRegression().fit(X, y)
        trends.append((district, model.coef_[0]))

trend_df = pd.DataFrame(trends, columns=['DISTRICT', 'Trend_Slope']).sort_values('Trend_Slope', ascending=False)
display(trend_df.head(10))

# Step 5: Time Series Plot for Top District
top_district = trend_df.iloc[0]['DISTRICT']
top_data = df[df['DISTRICT'] == top_district]

plt.figure(figsize=(8, 4))
plt.plot(top_data['YEAR'], top_data['TOTAL IPC CRIMES'], marker='o')
plt.title(f"Crime Trend in {top_district}")
plt.xlabel("Year")
plt.ylabel("Total IPC Crimes")
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 6: Export Results to CSV
stats_filtered = stats.dropna(subset=["TOTAL IPC CRIMES_std"])  # drop districts with only 1 year
stats_filtered.to_csv('/kaggle/working/district_stats.csv', index=False)
trend_df.to_csv('/kaggle/working/crime_trends.csv', index=False)
loadings.to_csv('/kaggle/working/pca_loadings.csv')

print("âœ… Exported cleaned district_stats.csv, crime_trends.csv, and pca_loadings.csv to /kaggle/working/")
```
